```python
import pandas as pd
import pycountry

# Read the CSV file into a DataFrame
df = pd.read_csv("GRUP 3-DP_LIVE.csv")


```

```python
df
```

```python
# Create a new column 'Country_Name'
df['Country_Name'] = None
```

```python
df
```

```python
# Iterate through the country code column and map the codes to full country names
for index, row in df.iterrows():
    try:
        country = pycountry.countries.get(alpha_3=row['LOCATION'])
        df.at[index, 'Country_Name'] = country.name
    except:
        # If the code is not found in the library, keep the cell empty
        pass

```

```python
df
```

```python
df.to_csv("countries.csv", index=False)
```
