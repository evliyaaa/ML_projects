## Read Data

```python
df = pd.read_csv("Advertising.csv")
df.sample(5)
```

## Polynomial Converter (Polinomial Çevirici)

```python
X = df.drop("sales", axis = 1)
y = df.sales
```

```python
poly_conv = PolynomialFeatures(degree = 2, include_bias = False)
```

```python
poly_conv.fit(X)
```

```python
poly_features = poly_conv.transform(X)
```

```python
poly_features.shape
```

```python
pd.DataFrame(poly_features, columns = ["TV", "Radio", "newspaper", "TV^2", "TV&Radio", "TV&Newspaper", "Radio^2", "Radio&Newspaper", "Newspaper^2"]).head()
```

### Train | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size = 0.3, random_state=101)
```

## Model Fit

```python
from sklearn.linear_model import LinearRegression
```

```python
model_poly = LinearRegression() #polynomial olarak kırılmış datayı linear modelde test ediyoruz
```

```python
model_poly.fit(X_train, y_train)
```

```python
y_pred = model_poly.predict(X_test)
```

```python
my_dict = {"Actual":y_test, "pred":y_pred, "residuals":y_test-y_pred}
compare = pd.DataFrame(my_dict)
compare.head(20)
```

```python
compare.head(20).plot(kind = "bar", figsize = (15, 9));
```

## Evaluation on the Test set

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
y_pred = model_poly.predict(X_test)
```

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    print("Model testing performance: ")
    print("-----------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
```

```python
eval_metric(y_test, y_pred)
```

```python
y_train_pred = model_poly.predict(X_train)
```

```python
eval_metric(y_train, y_train_pred)
```

## Let's find optimal degree of poly

```python
def poly(d):
    
    train_rmse_errors = []
    test_rmse_errors = []
    number_of_features = []
    
    for i in range(1, d):
        polynomial_converter = PolynomialFeatures(degree = i, include_bias =False)
        poly_features = polynomial_converter.fit_transform(X)
        X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
        model = LinearRegression(fit_intercept=True)
        model.fit(X_train, y_train)
        train_pred = model.predict(X_train)
        test_pred = model.predict(X_test)
        train_RMSE = np.sqrt(mean_squared_error(y_train,train_pred))
        test_RMSE = np.sqrt(mean_squared_error(y_test,test_pred))
        train_rmse_errors.append(train_RMSE)
        test_rmse_errors.append(test_RMSE)
        number_of_features.append(poly_features.shape[1])
    return pd.DataFrame({"train_rmse_errors": train_rmse_errors, "test_rmse_errors":test_rmse_errors, "number of features":number_of_features},
                        index=range(1,d))
```

```python
poly(10)
```

```python
plt.plot(range(1,10), poly(10)["train_rmse_errors"], label = "TRAIN")
plt.plot(range(1,10), poly(10)["test_rmse_errors"], label = "TEST")
plt.xlabel("Polynomial Complex")
plt.ylabel("RMSE")
plt.legend();
plt.show()
```

```python
plt.plot(range(1,6), poly(6)["train_rmse_errors"], label = "TRAIN")
plt.plot(range(1,6), poly(6)["test_rmse_errors"], label = "TEST")
plt.xlabel("Polynomial Complex")
plt.ylabel("RMSE")
plt.legend();
plt.show()
```
